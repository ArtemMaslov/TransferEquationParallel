```python
import pandas
import numpy
import matplotlib.pyplot as plt
import math
from IPython.display import Image
```

Программа численно решает задачу переноса:
$$
\begin{cases}
    \frac{\partial u (t, x)}{\partial t} + a \frac{\partial u(t, x)}{\partial x} = 0, 0 \le t \le T, 0 \le x \le X \\
    u(0, x) = A \sin(2 \pi x / T_x) \\
    u(t, 0) = -A \sin (2 \pi t / T_t)
\end{cases}
$$
Конечно-разностная схема Лакса-Вендроффа:
$$
\frac{u^{k+1}_m - u^k_m}{\tau} + a\frac{u^k_{m+1} - u^k_{m-1}}{2 h} - a^2\tau \frac{ u^{k}_{m+1} - 2 u^k_m + u^k_{m-1} }{2h^2} = f^k_m
$$
Аналитическое решение задачи Коши:
$$
u(t, x) = \sin (2 \pi \frac{x - a t}{T_x})
$$

### Сравнение решения с аналитическим

```python
data = pandas.read_csv("result.txt", sep = " ")
a = 1
PeriodX = 1

def Reference(x, t):
    return numpy.sin(2 * math.pi * (x - a * t) / PeriodX)

ref = Reference(data["x"], data["t"])
plt.plot(data["x"], data["u"], label = "mod")
plt.plot(data["x"], ref, label = "ref")
plt.legend()
plt.grid()
plt.show()

plt.plot(data["x"], numpy.log10(numpy.abs(ref - data["u"]) + 1e-6), label = "log|res|")
plt.legend()
plt.grid()
plt.show()
```
