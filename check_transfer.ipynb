```python
import pandas
import numpy
import matplotlib.pyplot as plt
import math
```

```python
data = pandas.read_csv("result.txt", sep = " ")
a = 1
PeriodX = 1

def Reference(x, t):
    return numpy.sin(2 * math.pi * (x - a * t) / PeriodX)

ref = Reference(data["x"], data["t"])
plt.plot(data["x"], data["u"], label = "mod")
plt.plot(data["x"], ref, label = "ref")
plt.legend()
plt.grid()
plt.show()

plt.plot(data["x"], numpy.log10(numpy.abs(ref - data["u"])), label = "log|res|")
plt.legend()
plt.grid()
plt.show()
```
